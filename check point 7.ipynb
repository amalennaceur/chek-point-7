```python
import pandas as pd 
data=pd.read_csv("CC GENERAL.csv", sep=",")
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
data['MINIMUM_PAYMENTS'].fillna(value=data['MINIMUM_PAYMENTS'].mean(),inplace=True)
data.isnull().sum().sum()
```

```python
data['CREDIT_LIMIT'].fillna(value=data['CREDIT_LIMIT'].mean(),inplace=True)
data.isnull().sum().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['CUST_ID']=encoder.fit_transform(data['CUST_ID'])
data
```

```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(data)  #Applying agglomerative algorithm with 5 clusters, using euclidean distance as a metric
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (data ['CUST_ID'] , data ["PURCHASES"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clutering")
ax.set_xlabel("CUST_ID")
ax.set_ylabel("PURCHASES)")
plt.colorbar(scatter)
```

```python
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (data ['CUST_ID'] , data ["BALANCE"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clutering")
ax.set_xlabel("CUST_ID")
ax.set_ylabel("BALANCE)")
plt.colorbar(scatter)
```

```python
import seaborn as sns
def plot_correlation_map( data ):

    corr = data.corr()

    s , ax = plt.subplots( figsize =( 24 , 20 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
plot_correlation_map(data)
```
